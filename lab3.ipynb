```python
t = np.arange(0,2.5,0.1)
y1 = np.sin(np.pi*t)
y2 = np.sin(np.pi*t+np.pi/2)
y3 = np.sin(np.pi*t-np.pi/2)
plt.xlabel('blue')
plt.ylabel('red')
plt.plot(t,y1,'r+',t,y2,'y*',t,y3,'g^')
```

```python
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)
plt.subplot(211)
plt.plot(t,y1,'bo')
plt.subplot(212)
plt.plot(t,y2,'ro')
```

```python
dx = 0.01
dy = 0.01
x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)
x,y = np.meshgrid(x,y)
def f(x,y):
    return(1- y**5+x**5)*np.exp(-x**2-y**2)
plt.contourf(x,y,f(x,y),8)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
dx = 0.01
dy = 0.01
x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)
x,y = np.meshgrid(x,y)
def f(x,y):
    return(1- y**5+x**5)*np.exp(-x**2-y**2)
ax.plot_surface(x,y,f(x,y))
```
